```python
import numpy as np
import matplotlib.pyplot as plt

t = [0, 30, 0.001]
def u(t):
 if t>=0 and t<2:
   return 0
 elif t>=2 and t<4:
   return-2
 elif t>=4 and t<10:
   return-8
 elif t>=10 and t<14:
   return 2
 elif t>=14 and t<20:
   return4
 elif t>=20 and t<22:
   return-2
 else:
   return-8
u = [u(ti) for ti in t]
moy = 0
eff = 0
nb_valeurs = len(u)
for i in range(nb_valeurs):
  moy += u[i]
  eff += u[i]**2
moy /= nb_valeurs
eff /= nb_valeurs
eff = np.sqrt(eff)
plt.plot(t, u)
plt.xlabel('t(ms)')
plt.ylabel('u(V)')
plt.ylim(-10,10)
plt.grid()
plt.show()

print(f'valeur moyenne: {moy:.2f} V')
print(f'valeur efficace: {eff:.2f}V')
print(t)

```
